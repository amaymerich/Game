##Introducción
En el desarrollo del presente trabajo, se busca modelar geométricamente una prótesis valvular cardíaca mecánica con base en el diseño de válvulas comerciales como la On-X®, reconocida por su eficacia clínica en el reemplazo de válvulas aórticas y mitrales. Este tipo de válvula se caracteriza por reproducir anatómicamente la forma del anillo aórtico natural y por incorporar un diseño optimizado que mejora la dinámica del flujo sanguíneo.

La válvula On-X® presenta una apertura angular de hasta 90°, cierre a 40° y una base en forma de campana, lo cual contribuye a favorecer el flujo laminar, reduciendo la regurgitación ,el retroceso de sangre por mal cierre, y la hemólisis,destrucción de glóbulos rojos debida a turbulencias o superficies inadecuadas.

##Hipótesis de modelado
Para el diseño tridimensional del prototipo de válvula cardíaca mecánica, se asume que la geometría está compuesta por tres elementos principales: un toroide que representa el anillo valvular, un cilindro hueco que actúa como estructura de soporte y canal de flujo, y un conjunto de hojas móviles modeladas como superficies cilíndricas curvas. Las valvas se consideran cuerpos rígidos que giran en torno a un eje fijo, sin deformarse, en un rango de apertura definido. El flujo sanguíneo que atraviesa la válvula se modela como un flujo laminar, unidireccional e ideal, sin turbulencias ni regurgitación. Se omiten fenómenos estructurales complejos como elasticidad, interacción fluido-estructura y rugosidad superficial. Estas simplificaciones permiten representar con precisión las características esenciales del funcionamiento valvular utilizando herramientas del Cálculo Vectorial, tales como funciones vectoriales, parametrizaciones y derivadas, facilitando así el análisis simbólico y la visualización computacional del modelo propuesto.

##Modelo geométrico propuesto
Con base en estas características, se propone un modelo geométrico tridimensional simplificado que mantenga las propiedades funcionales principales de la válvula, pero que sea viable de analizar y simular matemáticamente. El modelo se basa en las siguientes formas elementales:

Un **cilindro hueco**, que representa la estructura de soporte o conducto a través del cual circula la sangre.

Un **toroide**, que actúa como el anillo de sujeción de la válvula, ajustable a diámetros típicos entre 25 y 33 mm, de acuerdo con los estándares clínicos.

Un conjunto de hojas móviles o valvas, modeladas como **segmentos de superficies cilíndricas rotatorias**, con ángulos de apertura y cierre definidos (inspirados en los valores 90° y 40° de la válvula On-X®).

##Modelado matemático de las estructuras
Se modela mediante coordenadas cilíndricas.

ri: radio interno
re: radio externo
h: altura

##La parametrización vectorial es:

**Cilindro hueco**

Ecuación paramétrica

$$
\vec{r}(u,v) = \langle R \cos v,\ R \sin v,\ h \rangle
\quad \text{con}  \quad u \in [0, 1] , \quad v \in [0, 2\pi]
$$

**Toroide**




Ecuación paramétrica


$$
\vec{T}(u,v) = \left\langle \cos u \cdot (R + r \cos v),\ \sin u \cdot (R + r \cos v),\ r \sin v \right\rangle
\quad \text{con}  \quad u \in [0, 2\pi] , \quad v \in [0, 2\pi]
$$


**Hojas móviles**

Ecuación paramétrica

$$
\vec{H}(u, v) = \left\langle R \cos v,\ R \sin v,\ u \right\rangle
\quad ,\ v \in [0, \pi]
$$
